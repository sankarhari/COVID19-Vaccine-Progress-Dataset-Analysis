## Asking and Answering Questions

#### Q1: How India, United States, Brazil and Indonesia are performing in vaccinating the people from Year 2021 - 2022?

The purpose of the graph is to compare rate at which the various countries are vaccinating the their people. 

United States and United Kingdom as a wealthier country have fully vaccinated most of their population by May 2021 due to vaccine availability. Also that is the reason why the vaccination curve is flat after May 2021.

We can also notice other countries’ vaccination rate have increase much faster after May 2021. This may be due to weather countries have per-ordering the vaccination from the pharmaceutical companies for their population. So availability of vaccine to lesser wealthier country is getting affected.

In Indian’s and Brazil’s graph you have notice gap between People Vaccinated and People Fully Vaccinated population. The may be due to countries changing their vaccination interval between Dose 1 and Dose 2 based on availability of vaccines. They have targeted to vaccinate their population with at-least one dose first. So they have increase the vaccination interval.

But in US and UK, due to availability of vaccine they have set the vaccination interval lesser when compared to other countries.

In this graph X-Axis indicates Date, Y-Axis indicates number of people vaccinated.

```python
temp_compare_list_count = ["United Kingdom","United States","India","Brazil","Indonesia"]

figure, axis = plt.subplots(len(temp_compare_list_count), 1, figsize=(40,40))

axis_counter = 0
for each_count in temp_compare_list_count:
    covid_vaccin_df_set = covid_vaccin_df[covid_vaccin_df["country"] == each_count].sort_values("date")
    temp_mean = covid_vaccin_df_set.mean(numeric_only = True)
    ld_date = covid_vaccin_df_set["date"]
    ld_total_vaccinations=covid_vaccin_df_set["total_vaccinations"]
    ld_people_vaccinated=covid_vaccin_df_set["people_vaccinated"]
    ld_people_fully_vaccinated=covid_vaccin_df_set["people_fully_vaccinated"]
    
    axis[axis_counter].plot(ld_date,ld_total_vaccinations,label = "Total Vaccinations")
    axis[axis_counter].plot(ld_date,ld_people_vaccinated,label = 'People Vaccinated')
    axis[axis_counter].plot(ld_date,ld_people_fully_vaccinated,label = 'People Fully Vaccinated')
    axis[axis_counter].set_title("Vaccination Rate in "+each_count)
    axis[axis_counter].set_ylabel("People Count")
    axis[axis_counter].set_xlabel("Date")
    axis[axis_counter].legend()
    axis_counter += 1
```

#### Q2 - What are the top 10 countries who have done highest vaccinations globally?

This Bar Graph shows top 10 countries who have done highest number vaccine doses. One of the main factor of this statistic is countries population. As we all know China and India as the highly populated country they have procured and vaccinated the height number of population globally. 

Countries like United States, Brazil, Indonesia, Japan, Pakistan, Bangladesh,Vietnam and Mexico are the countries which have done highest number of vaccination followed by China and India in this particular order.
 
In this graph X-Axis indicates top 10 who have done highest vaccinations globally. Y-Axis indicates the Population Vaccinated. 

```python
covid_vaccin_df = covid_vaccin_df.sort_values("total_vaccinations",ascending=False)
countries = covid_vaccin_df["country"].unique()
countries_lite = countries[0:10]
stat_lite = []
np.append(stat_lite,123)
for each_count in countries[0:10]:
    temp_var = covid_vaccin_df[covid_vaccin_df["country"] == each_count].max()
    temp_stat = temp_var["total_vaccinations"]
    stat_lite.append(temp_stat)
    
stat_lite = np.array(stat_lite)
stat_lite

plt.figure(figsize=(20,6))
plt.bar(countries_lite, stat_lite)
 
plt.xlabel("Country")
plt.ylabel("Population Vaccinated")  
plt.title("Top 10 countries who have done highest vaccinations globally")
plt.show()
```

#### Q3 - What are percentage of various Vaccines used by each countries globally except Pfizer/BioNTech?

I have percentage stacked bar chart to show percentage of other vaccines procured by each country other than Pfizer. 

X-Axis represents country name
Y-Axis represents Vaccine Count

Moderna and Oxford/AstaZeneca are the two other vaccines used widely in Italy, France, Germany, United States and European Union 

```python
country_list = covid_vaccin_man_df["location"].unique()
# country_list = country_list[country_list != "United States"]
# country_list = country_list[country_list != "European Union"]
vaccine_list = covid_vaccin_man_df["vaccine"].unique()
vaccine_map = {}
for each_country in country_list:
    for each_vaccine in vaccine_list:
        max_vaccine_temp = 0
        temp_Data_df = covid_vaccin_man_df[covid_vaccin_man_df["location"] == each_country]
        temp_Data_df = temp_Data_df[temp_Data_df["vaccine"] == each_vaccine]
        max_vaccine_temp = temp_Data_df["total_vaccinations"].max()
        if(math.isnan(max_vaccine_temp)):
            max_vaccine_temp = 0
        if each_vaccine in vaccine_map:
            temp_list = vaccine_map[each_vaccine]
            temp_list.append(max_vaccine_temp)
            vaccine_map[each_vaccine] = temp_list
        else:
            vaccine_map[each_vaccine] = [max_vaccine_temp]
#         print("Country Name: " + each_country + "\nVaccine Name: " + each_vaccine + "\nMax: " + format(max_vaccine_temp))
#         print("\n--------------------")
#     print("\n=====================")

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots(1,1,figsize=(40,15))
vaccine_list = vaccine_list[vaccine_list != "Pfizer/BioNTech"]
for each_vaccine in vaccine_list:
    ax.bar(country_list, vaccine_map[each_vaccine], width, label=each_vaccine)

ax.set_ylabel('Vaccinations Count')
ax.set_xlabel('Countries')
ax.set_title('Countrie\'s Vaccination Count Based on Manufactorers')
ax.legend()

plt.show()
```

#### Q4 - What are percentage of various Vaccines used by European countries?

This bar graph is focused on some major European counties which shows percentage of various vaccination procured by counties to vaccinate their population.

We can notice these European countries have procured more COVID-19 vaccine developed by American. Especially vaccine developed by Pfizer/BioNTech, Around 80% of vaccines procured are Pfizer.

Secondly Moderna is the next height percentage of vaccine produced by other European countries except Romania. Romania have procured more Johnson&Johnson vaccine than Moderna unlike other European counties.

Though Johnson&Johnson is developed in Netherlands they have preferred US developed vaccine due to high efficacy in preventing severe COVID-19.

```python
country_list = covid_vaccin_man_df["location"].unique()
vaccine_list = covid_vaccin_man_df["vaccine"].unique()
country_list =["Austria","Belgium","France","Finland","Germany","Italy","France","Romania","Spain"]
figure, axis = plt.subplots(3, 3, figsize=(60,20))

fig_x_axis = 0
fig_y_axis = 0
fig_counter = 1
for each_country in country_list:
    temp_vaccine = []
    for each_vaccine in vaccine_list:
        temp_Data_df = covid_vaccin_man_df[covid_vaccin_man_df["location"] == each_country]
        temp_Data_df = temp_Data_df[temp_Data_df["vaccine"] == each_vaccine]
        if temp_Data_df.empty:
            temp_vaccine.append(0)
        else:
            temp_vaccine.append(temp_Data_df["total_vaccinations"].max())
    #     brea
#     print(temp_vaccine)
    axis[fig_x_axis,fig_y_axis].bar(vaccine_list, temp_vaccine)
    axis[fig_x_axis,fig_y_axis].set_title(each_country)
    axis[fig_x_axis,fig_y_axis].set_xlabel("Vaccine Name")
    axis[fig_x_axis,fig_y_axis].set_ylabel("Dosage Count")
#     print(each_country,fig_x_axis,fig_y_axis)
    if fig_y_axis > 1:
        fig_y_axis = 0
        fig_x_axis += 1
    else:
        fig_y_axis += 1
    fig_counter += 1
```

#### Q5 - What Vaccine have contributed the most to handle the COVID19 pandemic?

This Pie Chart shown what percentage of vaccines are used most by all counties globally.

1. Pfizer/BioNTech
2. Moderna
3. Oxford/AstaZeneca
4. Johnson&Johnson
5. Sinovac
6. Sinopharm/Beijing

these are the most procured vaccine by all counties globally in this particular order. United States developer Pfizer and Moderna used widely globally. This might be due to high efficacy in preventing COVID-19 and product rate. Other vaccines developed in UK, Netherlands, China where used predominantly used within their countries and sent to near by countries. In this data-set some of the vaccines where not listed. Especially CoviSheld, Which is widely used in India which is developed by Serum Institute of India Pvt Ltd.  

```python

country_list = covid_vaccin_man_df["location"].unique()
vaccine_list = covid_vaccin_man_df["vaccine"].unique()

vaccine_sum_list = []
for each_vaccine in vaccine_list:
    Total_Count = 0
    for each_country in country_list:
        temp_Data_df = covid_vaccin_man_df[covid_vaccin_man_df["location"] == each_country]
        temp_Data_df = temp_Data_df[temp_Data_df["vaccine"] == each_vaccine]
        max_vaccine_temp = temp_Data_df["total_vaccinations"].max()
        if(math.isnan(max_vaccine_temp)):
            max_vaccine_temp = 0
        Total_Count += max_vaccine_temp
#         print("Country Name: " + each_country + "\nVaccine Name: " + each_vaccine + "\nMax: " + format(max_vaccine_temp))
#         print("\n--------------------")
    vaccine_sum_list.append(Total_Count)
# print(vaccine_list)
# print(vaccine_sum_list)
figure, axis = plt.subplots(1, 1, figsize=(20,20))
axis.pie(vaccine_sum_list, labels = vaccine_list, autopct='%1.1f%%')
axis.legend(title = "Vaccine Name")
axis.set_title("Percentage of vaccine used globally")
```
